```python
#web scrapping
import requests
from bs4 import BeautifulSoup

#translation
from googletrans import Translator, constants

#summarization
import spacy
from spacy.lang.en.stop_words import STOP_WORDS
from string import punctuation
from collections import Counter
from heapq import nlargest
from rouge import Rouge

def calculate_rouge_scores(hypothesis, reference):
    rouge = Rouge()
    scores = rouge.get_scores(hypothesis, reference)
    return scores
```

# **WEB SCRAPPING**

```python
r = requests.get('https://www.divyabhaskar.co.in/utility/automobile/news/this-suv-will-also-run-on-e-20-fuel-the-information-was-leaked-before-the-launch-131086628.html')

# check status code for response received
# success code - 200
print(r)

# print content of request
print(r.content)
```

```python
soup = BeautifulSoup(r.content, 'html.parser')
titlesHTML=soup.find_all('h1')
content = soup.find_all('p')

print(titlesHTML)
print(content)
```

```python
titles=[]
for line in titlesHTML:
  titles.append(line.text)
titles
```

```python
text=[]

for line in content:
  text.append(line.text)

text
```

# **TRANSLATION**

```python
translator = Translator()
```

```python
for t in titles:
  translation = translator.translate(t)
  trans_title = translation.text

trans_title
```

```python
trans_text=[]

for t in text:
  translation = translator.translate(t)
  trans_text.append(translation.text)

text_final=",".join(trans_text)
text_final
```

# **EXTRACTIVE SUMMARIZATION**

---



```python
import spacy
from heapq import nlargest

# Load spaCy English model
nlp = spacy.load("en_core_web_sm")



# Tokenize text into sentences using spaCy
doc = nlp(text_final)

# Extract named entities from the text
entities = set([ent.text.lower() for ent in doc.ents])

# Calculate sentence scores based on the presence of named entities
sentence_scores = {}
for sent in doc.sents:
    score = sum(1 for token in sent if token.text.lower() in entities)
    sentence_scores[sent.text] = score

# Select top sentences based on scores
summarized_sentences = nlargest(8, sentence_scores, key=sentence_scores.get)
summary = ' '.join(summarized_sentences)
print(summary)
```

```python

rouge_scores = calculate_rouge_scores(summary, text_final)

print("ROUGE scores:")
print(rouge_scores)
```
